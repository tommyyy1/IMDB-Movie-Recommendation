```python
# Scroll down and click "50 more" multiple times
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd

driver=webdriver.Chrome()
# Open the IMDb action movie 2024 page
driver.get("https://www.imdb.com/search/title/?title_type=feature&genres=action&release_date=2024-01-01,2024")
time.sleep(5)

while True:
    try:
        # Find the "50 more" button and scroll to it
        load_more_button = driver.find_element(By.XPATH, "//span[contains(text(),'50 more')]")
        driver.execute_script("arguments[0].scrollIntoView();", load_more_button)
        time.sleep(2)  # Wait for the button to become interactable
        
        # Click the button
        load_more_button.click()
        print("Clicked '50 more' button...")
        time.sleep(5)  # Wait for new movies to load
    except Exception:
        print("No more '50 more' button found. All movies are loaded.")
        break  # Exit loop when all movies are loaded

# Extract movie names
movie_titles = driver.find_elements(By.CSS_SELECTOR, '.ipc-title.ipc-title--base.ipc-title--title.ipc-title-link-no-icon.ipc-title--on-textPrimary')
movie_names = [title.text for title in movie_titles]

# Extract movie descriptions
movie_descriptions = driver.find_elements(By.XPATH, "//div[@class='ipc-html-content-inner-div']")
movie_des = [desc.text for desc in movie_descriptions]

# Match lengths: Keep only pairs where both title and description exist
min_length = min(len(movie_names), len(movie_des))
movie_names = movie_names[:min_length]
movie_des = movie_des[:min_length]

print(f" Extracted {len(movie_names)} movie titles and {len(movie_des)} descriptions.")

# Save to CSV
df = pd.DataFrame({"Movie Name": movie_names, "Storyline": movie_des})
df.to_csv("IMDb_2024_Action_Movies.csv", index=False)
print("Data saved successfully to 'IMDb_2024_Action_Movies.csv'.")

# Close browser
driver.quit()
```

```python
#Text cleaning

import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Download required NLTK resources
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')

# Load CSV File
df = pd.read_csv("modified_movies.csv")  # Ensure the CSV has columns: "title" and "description"

# Initialize Lemmatizer
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Function to clean text
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\s+', ' ', text)  # # Ensure single spaces
    text = re.sub(r'[^\w\s:]', '', text)    # Remove special characters but keep colons
    tokens = word_tokenize(text)  # Tokenization
    tokens = [word for word in tokens if word not in stop_words]  # Remove stopwords
    tokens = [lemmatizer.lemmatize(word) for word in tokens]  # Lemmatization
    return " ".join(tokens)

# Apply Cleaning to Movie Descriptions
df['cleaned_title'] = df['Movie Name'].apply(clean_text)
df['cleaned_storyline'] = df['Storyline'].apply(clean_text)

# Save Cleaned Data
df.to_csv("cleaned_movies.csv", index=False)

# Display Sample
df.head(5)

```

```python
%%writefile movieapp.py
import streamlit as st
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset (Ensure the CSV has 'title' and 'cleaned_description' columns)
@st.cache_data
def load_data():
    df = pd.read_csv("cleaned_movies.csv")
    return df

df = load_data()

# Apply TF-IDF Vectorization
@st.cache_resource
def compute_tfidf():
    vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = vectorizer.fit_transform(df['cleaned_storyline'])
    similarity_matrix = cosine_similarity(tfidf_matrix)
    return vectorizer, similarity_matrix

vectorizer, similarity_matrix = compute_tfidf()

# Function to get movie recommendations
def recommend_movies(movie_title, top_n=5):
    if movie_title not in df['cleaned_title'].values:
        return ["Movie not found in database."]

    idx = df[df['cleaned_title'] == movie_title].index[0]
    sim_scores = list(enumerate(similarity_matrix[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:top_n+1]
    recommended_titles = [df.iloc[i[0]]['cleaned_title'] for i in sim_scores]
    recommended_storylines = [df.iloc[i[0]]['cleaned_storyline'] for i in sim_scores]
    
    return recommended_titles, recommended_storylines

# Streamlit UI
st.title("🎬 IMDb 2024 Movie Recommendation System")

# Movie Selection Dropdown
selected_movie = st.selectbox("Select a Movie:", df['cleaned_title'].values)

if st.button("Get Recommendations"):
    recommendations, descriptions = recommend_movies(selected_movie)

    st.subheader("Recommended Movies:")

    # Create columns to display movie titles and storylines side by side
    cols = st.columns(2)  # Create two columns
    
    with cols[0]:
        st.subheader("Movie Title")
        for movie in recommendations:
            st.write(f"📌 {movie}")

    with cols[1]:
        st.subheader("Storyline")
        for desc in descriptions:
            st.write(f"📜 {desc[:250]}...")  # Display first 250 characters of the description
        

# Show Data Preview (Optional)
if st.checkbox("Show Movie Dataset"):
    st.write(df[['cleaned_title', 'cleaned_storyline']].head(10))
```

```python
%%writefile movieapp1.py

import streamlit as st
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset (Ensure the CSV has 'title' and 'cleaned_description' columns)
@st.cache_data
def load_data():
    df = pd.read_csv("cleaned_movies.csv")
    return df

df = load_data()

# Apply TF-IDF Vectorization
@st.cache_resource
def compute_tfidf():
    vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = vectorizer.fit_transform(df['cleaned_storyline'])
    return vectorizer,  tfidf_matrix

vectorizer, tfidf_matrix = compute_tfidf()
# Calculate similarity matrix outside the function
similarity_matrix = cosine_similarity(tfidf_matrix)

# Function to get movie recommendations based on user input
def recommend_movies(user_input, top_n=5):
    # Process the user input using the same vectorizer
    user_input_tfidf = vectorizer.transform([user_input])

# Calculate cosine similarity between input and all movie descriptions
    sim_scores = cosine_similarity(user_input_tfidf, tfidf_matrix).flatten()
    
    # Get the top n most similar movies
    sorted_indices = sim_scores.argsort()[-top_n:][::-1]
    
   
    # Get recommended movie titles and descriptions
    recommended_titles = df.iloc[sorted_indices]['cleaned_title'].values
    recommended_descriptions = df.iloc[sorted_indices]['cleaned_storyline'].values

    return recommended_titles, recommended_descriptions

# Streamlit UI
st.title("🎬 IMDb 2024 Movie Recommendation System")

# User Input for Movie Storyline
user_input = st.text_area("Enter a Movie Storyline (short description or plot):", height=200)

# Display recommendations when button is pressed
if st.button("Get Recommendations"):
    if user_input.strip():
        # Get the top 5 recommended movies based on the user input
        recommended_titles, recommended_descriptions = recommend_movies(user_input)

        st.subheader("Top 5 Recommended Movies:")
        for i in range(5):
            st.write(f"**{i + 1}. {recommended_titles[i]}**")
            st.write(f"   *Description:* {recommended_descriptions[i]}")
    else:
        st.warning("Please enter a movie storyline to get recommendations.")

# Show Data Preview (Optional)
if st.checkbox("Show Movie Dataset"):
    st.write(df[['cleaned_title', 'cleaned_storyline']].head(10))
```
